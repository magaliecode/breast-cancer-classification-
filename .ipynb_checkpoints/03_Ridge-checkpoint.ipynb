In medical classification tasks like breast cancer detection, **recall is critical** — missing a malignant case is far more costly than raising false alarms. While our baseline Logistic Regression performs decently, we turn to **Ridge (L2) regularization** to enhance robustness and generalization.

---

### Why Use Ridge?

- **Preserves All Features**: Ridge shrinks coefficients without discarding features, which is ideal in biomedical datasets where many predictors may carry important signals.
- **Handles Multicollinearity**: L2 regularization stabilizes models when predictors are highly correlated by distributing weights across them, rather than eliminating one arbitrarily (as Lasso might).
- **Reduces Overfitting**: Adding an L2 penalty discourages overly large coefficients, especially helpful in small datasets where overfitting is a concern.

---

### How It Works

- Adds an L2 penalty term to the loss function.
- The strength of regularization is controlled by **`C`** (inverse of regularization strength); **smaller `C` → stronger regularization**.
- Coefficients are pulled closer to zero, but none are eliminated.

---

### What We're Doing

- Preprocess the data differently : instead of dropping `HOMA` like we did previously for baseline logistic regression, we use all features because Ridge allows us to retain it. 
- Use **scaled features**
- Train a Logistic Regression model with **L2 penalty**
- Perform **cross-validation** to tune `C`, optimizing for **recall**
- Evaluate using:
  - Accuracy, Precision, Recall, F1 Score
  - ROC AUC
  - ROC Curve and Confusion Matrix 
  comparing different thresholds 

This approach balances **model interpretability** with **high sensitivity**, making it suitable for clinical decision support.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, RocCurveDisplay
```

```python
data=pd.read_csv("dataR2.csv")
data["target"] = data["Classification"].map({1: 0, 2: 1})
```

```python
skewed_features = ["Insulin", "Leptin", "Resistin", "MCP.1", "Adiponectin", "HOMA"] 

# Apply log(1+x) transformation to handle skewness & avoid log(0)
for col in skewed_features:
    data[col] = np.log1p(data[col])

data[skewed_features].hist(bins=20, figsize=(12, 8), edgecolor='black')
plt.suptitle("Skewed Feature Distributions", fontsize=16)
plt.show()
```

```python
X = data.drop(columns=["target", "Classification"])
y = data["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Ridge Logistic Regression (with CV tuned for Recall)
from sklearn.model_selection import GridSearchCV, StratifiedKFold


# Model & CV setup
ridge = LogisticRegression(
    penalty="l2",
    solver="liblinear",    
    class_weight="balanced",
    max_iter=1000,
    random_state=42
)

param_grid = {
    "C": np.logspace(-4, 4, 20)  # search from strong to weak regularization
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid = GridSearchCV(
    estimator=ridge,
    param_grid=param_grid,
    scoring="recall",      # prioritize catching positives
    cv=cv,
    n_jobs=-1,
    refit=True,
    verbose=0
)
```

```python
# Fit
grid.fit(X_train_scaled, y_train)

best_ridge = grid.best_estimator_
print(f"Best C (by recall): {grid.best_params_['C']:.5f}")
```

```python
y_proba = best_ridge.predict_proba(X_test_scaled)[:, 1]


thresholds = [0.5, 0.4, 0.3, 0.2, 0.1]

results = []

for thresh in thresholds:
    y_pred = (y_proba >= thresh).astype(int)
    results.append({
        'Threshold': thresh,
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'F1 Score': f1_score(y_test, y_pred)
    })

df_results = pd.DataFrame(results)

print(df_results.to_string(index=False))


auc  = roc_auc_score(y_test, y_proba)
print(f"ROC AUC:   {auc:.3f}")
```

### Threshold Tuning: Balancing Sensitivity and Specificity

In medical diagnosis — especially in breast cancer detection — **recall (sensitivity)** is the most critical metric. A false negative (i.e., missing a malignant case) can have serious consequences. To address this, we evaluated the model at multiple decision thresholds to find the best balance between **recall**, **precision**, and overall performance.

| Threshold | Accuracy | Precision | Recall | F1 Score |
|-----------|----------|-----------|--------|----------|
| 0.5       | 0.708    | 0.800     | 0.615  | 0.696    |
| 0.4       | 0.750    | 0.769     | 0.769  | 0.769    |
| **0.3**   | 0.750    | 0.706     | **0.923** | **0.800** |
| 0.2       | 0.625    | 0.600     | **0.923** | 0.727    |
| 0.1       | 0.542    | 0.542     | **1.000** | 0.703    |

- **Threshold = 0.5 (default)**: Recall is too low (0.615), despite high precision. This is **not acceptable** in a medical setting.
- **Threshold = 0.4**: Improves recall and maintains balance — a reasonable compromise.
- **Threshold = 0.3**: Recall jumps to **0.923** with a still-decent precision (0.706) and F1 score (0.800). This is the **best trade-off**, ensuring high sensitivity without excessive false positives.
- **Thresholds 0.2 and 0.1**: Push recall to extremes (up to 1.0), but at the cost of **precision** and overall accuracy. These would likely cause **too many false alarms**, reducing clinical usability.

The **ROC AUC of 0.804** indicates good overall discrimination ability, reinforcing that threshold adjustment is key to aligning model behavior with medical priorities.

**Final choice: Threshold = 0.3**, as it ensures high recall with acceptable precision — optimal for early cancer detection.

The plot above illustrates how precision and recall vary with different decision thresholds.

- As expected, **lowering the threshold increases recall** — we become more sensitive to identifying positive (malignant) cases.
- However, this comes at the cost of **reduced precision** — more false positives are introduced.

This visualization supports our earlier decision to set the classification threshold to **0.3**, prioritizing **high sensitivity (recall)** while keeping **precision and false positive rate reasonable**.
