# 6. Numerical features Distribution 

## Overall distribution 
We examine distributions to detect skewness and outliers, which may influence model performance. 

```python
features = ["Age", "BMI", "Glucose", "Insulin", "HOMA", "Leptin", "Adiponectin", "Resistin", "MCP.1"]

data[features].hist(bins=20, figsize=(15, 10), edgecolor='black')
plt.suptitle("Numerical Feature Distributions", fontsize=16)
plt.show()
```

- __Age__ and __BMI__ are approximately normally distributed with moderate spread, indicating balanced representation across ranges.

- __Glucose__ is slightly right-skewed, with most values clustered between 80–120 but a few very high values (~200).

- __Insulin__, __HOMA__, __Leptin__, __Resistin__, and __MCP.1__ show strong right skewness, with many small values and a few extreme large values.

- __Adiponectin__ is moderately right-skewed, with most values between 5–15 but some large outliers.

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(features, 1):
    plt.subplot(3, 3, i)  # 3 rows, 3 columns
    sns.boxplot(x=data[col], color="skyblue")
plt.tight_layout()
plt.show()
```

The boxplots confirm the presence of outliers in several features:

- Minimal outliers: Age, BMI.

- Moderate outliers: Glucose, Leptin, MCP.1.

- Many outliers: Insulin, HOMA, Adiponectin, Resistin — likely due to their highly skewed distributions.

- These extreme values can influence model performance if not treated (e.g., log transformation, robust scaling, or outlier removal).

### Summary
The dataset contains features with very different distribution shapes — some close to normal (Age, BMI), some moderately skewed (Glucose, Adiponectin, MCP.1), and some heavily skewed with long tails (Insulin, HOMA, Leptin, Resistin). Outliers are present in nearly all features, especially in those with skewed distributions.

Implications for preprocessing:
- Consider transformations (e.g., log or Box-Cox) for skewed variables.

- Apply scaling (StandardScaler or RobustScaler) before model training.

- Investigate extreme outliers to decide whether they should be capped, removed, or kept depending on domain context.

# 7. Correlation Analysis 

In this section, we will use both Pearson correlation and Spearman correlation for the correlation analysis.
We use Pearson correlation to measure linear relationships, and Spearman correlation to measure monotonic relationships (which can be non-linear).
Including both helps capture relationships that might not be strictly linear — for example, a variable could have a weak Pearson correlation but a strong Spearman correlation if its association with the target is non-linear but still consistently increasing or decreasing.

### Pearson Correlation (Linear Relationships)

```python
corr = data[features + ["target"]].corr(method="pearson")
mask = np.triu(np.ones_like(corr, dtype=bool))
plt.figure(figsize=(8, 6))
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f",
            cmap="coolwarm", vmin=-1, vmax=1, linewidths=.5)
plt.title("Correlation Heatmap (Pearson)")
plt.show()
```

```python
target_corr = corr["target"].drop("target").sort_values(ascending=False)
display(target_corr.to_frame("corr_with_target"))
```

- __Glucose__ shows the strongest positive correlation with the target (0.384), followed by __HOMA__ (0.284) and __Insulin__ (0.277).

- __Resistin__ has a weaker but still noticeable positive correlation (0.227).

- __BMI__ (-0.133), __Age__ (-0.044), __Adiponectin__ (-0.020), and __Lepti__n (-0.001) show little to no linear correlation with the target.

Feature-to-feature:

- __HOMA__ and __Insulin__ are extremely correlated (0.93) — strong multicollinearity risk.

- __HOMA__ and __Glucose__ (0.70) also have high correlation.

- __Leptin__ and __BMI__ moderately correlated (0.57).

```python
corr_s = data[features + ["target"]].corr(method="spearman")
plt.figure(figsize=(8, 6))
sns.heatmap(corr_s, mask=mask, annot=True, fmt=".2f",
            cmap="coolwarm", vmin=-1, vmax=1, linewidths=.5)
plt.title("Correlation Heatmap (Spearman)")
plt.show()

target_corr_s = corr_s["target"].drop("target").sort_values(ascending=False)
display(target_corr_s.to_frame("spearman_with_target"))
```

- __Glucose__ remains the top feature (0.456) with an even stronger monotonic correlation compared to Pearson — indicating a potentially non-linear relationship.

- __Resistin__ also shows stronger monotonic correlation (0.290) than linear (0.227).

- __HOMA__ is stable across both methods (~0.278).

- __Insulin__ shows a drop in monotonic correlation (0.207) vs. linear (0.277), suggesting its relationship with the target is more linear.

- Low correlations remain for __MCP.1__, __Adiponectin__, __Leptin__, __Age__, and __BMI__.

Feature-to-feature:

- __HOMA–Insulin__ correlation rises to 0.98.

- __Leptin–BMI__ increases to 0.66.


### Summary 
- __Glucose__ is the most important predictor in both metrics, with a stronger Spearman correlation — non-linear models (e.g., tree-based, kernel methods) may capture this better.

- __Resistin__ gains importance under Spearman, suggesting monotonic but non-linear patterns.

- __Insulin__ is more relevant for purely linear models, but is highly collinear with __HOMA__.

- __Leptin__, __Adiponectin__, __Age__, __BMI__ show weak associations individually but may still be useful in interaction terms or non-linear models.

- Multicollinearity between __HOMA__, __Insulin__, and __Glucose__ should be addressed before modeling (e.g., using VIF analysis or feature selection).

# 8. Multicollinearity analysis 

```python
from sklearn.linear_model import LinearRegression
```

```python
X = data[features].values

vifs = []
for i in range(X.shape[1]):
    y = X[:, i]
    X_others = np.delete(X, i, axis=1)
    model = LinearRegression().fit(X_others, y)
    r2 = model.score(X_others, y)
    vif = 1.0 / max(1.0 - r2, 1e-12)  # numerical safety
    vifs.append(vif)

vif_df = pd.DataFrame({'Feature': features, 'VIF': vifs}).sort_values('VIF', ascending=False)
print(vif_df)
```

### Multicollinearity Analysis (VIF)

Variance Inflation Factor (VIF) measures how much the variance of a regression coefficient is inflated due to multicollinearity with other features.

**Interpretation guidelines:**
- VIF ≈ 1 → No multicollinearity
- VIF 1–5 → Moderate multicollinearity (usually acceptable)
- VIF > 5 (or >10) → High multicollinearity → may cause instability in linear models


**Results:**
- **HOMA** (18.2) and **Insulin** (12.45) show very high multicollinearity, confirming the strong correlations seen earlier (HOMA is calculated from Glucose and Insulin).
- **Glucose** has moderate multicollinearity (2.93).
- All other features have low VIF (<2), indicating minimal collinearity issues.

**Implications for modeling:**
- High VIF features may cause instability in models that assume independent predictors (e.g., Logistic Regression without regularization).
- Options:
  - Drop one of the highly collinear variables (e.g., remove HOMA and keep Glucose & Insulin)
  - Use dimensionality reduction (PCA)
  - Use regularized models (Ridge, Lasso, ElasticNet) or tree-based algorithms which are less sensitive to multicollinearity


# Next Steps


Our EDA revealed three important data characteristics that influence preprocessing and model choice:

1. **Multicollinearity** – Features such as **HOMA** and **Insulin** are highly correlated (VIF > 10), which can cause instability in linear models.
2. **Right skewness** – Variables such as **Insulin**, **HOMA**, and **Resistin** are heavily right-skewed, which can affect models that assume normally distributed features.
3. **Outliers** – These same features also contain extreme values, which can distort coefficients in linear models and distance-based algorithms.

### Modeling plan
We will train and compare **4 different algorithms** from different model families:

1. **Logistic Regression (baseline)** – Interpretable, but sensitive to multicollinearity, skewness, and outliers.
2. **Regularized Logistic Regression (Ridge or Lasso)** – Reduces the impact of multicollinearity and may handle some noise, but still affected by skewness/outliers.
3. **Tree-based model (Random Forest or XGBoost)** – Captures non-linear relationships, less sensitive to multicollinearity, skewness, and outliers.
4. **Support Vector Machine (SVM)** – Can capture complex decision boundaries, but requires scaling and is sensitive to outliers.

### Model-specific preprocessing

| Model | Multicollinearity | Skewness | Outliers | Scaling |
|-------|------------------|----------|----------|---------|
| Logistic Regression | Drop correlated vars or regularize | Apply log/Box-Cox transform | Treat/remove extreme outliers | Required |
| Regularized Logistic Regression | Keep all vars | Apply log/Box-Cox transform | Treat/remove extreme outliers | Required |
| Tree-based models | Keep all vars | Transformation optional | Outliers less impactful | Not required |
| SVM | Keep all vars | Apply log/Box-Cox transform | Treat/remove extreme outliers | Required |

### Why multiple models?
Using several models allows us to:
- Compare predictive performance across different algorithm families.
- See how skewness and outliers affect each model differently.
- Balance interpretability (Logistic Regression) and predictive power (Tree-based models).

### Goal
By tailoring preprocessing to each model’s sensitivity to **multicollinearity**, **skewness**, and **outliers**, we ensure that:
- Every algorithm receives the data in its optimal form.
- Our comparison is fair and insightful.
- The final recommendations are based on evidence, not just one algorithm’s behavior.
